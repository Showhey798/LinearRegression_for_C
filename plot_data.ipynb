```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
"""
    トイデータの作成(sin関数に正規分布のノイズを追加したもの)
"""
def generate_data(size, min=0,max=1, scale=0.3):
    x = np.linspace(min, max, size)
    y = np.sin(2*np.pi*x) + np.random.normal(0, scale, size)
    return x, y

""" デザイン行列の作成 """
def design_matrix(x, m):
    phi = np.ones((len(x), m+1))
    for i in np.arange(1, m+1):
        phi[:, i] = x**i
    return phi
```

```python
""" データの生成、書き込み """
x, y = generate_data(size=10)
df = pd.DataFrame([x, y]).T
df.to_csv("./csv_files/toy_data.csv", index=False, header=None)
```

```python
""" データを受け取りグラフを出力 """
plt.scatter(df[0], df[1], facecolor="none", edgecolor="red")

w = np.array(pd.read_csv("./csv_files/regular_least_coef.csv", header=None))
x = np.linspace(0, 1, 1000)
phi = design_matrix(x, 9)

plt.plot(x, phi@w, color="green")

plt.plot(x, np.sin(2*np.pi*x), color="blue")

plt.show()
```
